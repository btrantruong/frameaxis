```python
%%time
sa = SemAxis(CoreUtil.load_embedding("../../data/embeddings/glove.840B.300d.w2vformat.txt", is_binary=False), 
               axes_str=CoreUtil.load_wordnet_antonyms_axes())
```

# big table for bias

```python
COLUMNS = [str(c) for c in sorted(sa.axes.keys()) if len(c) == 2]

with open("big_table_by_average.tsv", "w") as fo_a, open("big_table_by_kurtosis.tsv", "w") as fo_k:
    fo_a.write("genre\trating\treview\t{}\n".format("\t".join(COLUMNS)))
    fo_k.write("genre\trating\treview\t{}\n".format("\t".join(COLUMNS)))
    for loop_index, (row_index, row) in enumerate(df.iterrows()):
        if loop_index % (1000) == 0:
            logger.info(loop_index)
        try:
            _, mean, kurtosis = sa.compute_document_mean_kurtosis_with_tf([row['review']], min_freq = 1)
        except ValueError as ve:
#             logger.exception("no vocab")
            continue
        except:
            logger.exception("???")
            continue

        fo_a.write("{}\t{}\t{}\t{}\n".format(row['genre'], row['rating'], row['review'], "\t".join([str(v) for v in mean])))
        fo_k.write("{}\t{}\t{}\t{}\n".format(row['genre'], row['rating'], row['review'], "\t".join([str(v) for v in kurtosis])))
```

# big table for intensity

```python
COLUMNS = [str(c) for c in sorted(sa.axes.keys()) if len(c) == 2]

with open("big_table_by_second_moment_with_corpus_mean.tsv", "w") as fo_k:
    fo_k.write("genre\trating\treview\t{}\n".format("\t".join(COLUMNS)))
    df_corpus = pd.read_csv("big_table_by_average.tsv", sep="\t")
    corpus_mean = np.mean(df_corpus.drop(
        columns=[c for c in df_corpus.columns if '(' not in c]).values, axis=0)
    for loop_index, (row_index, row) in enumerate(df.iterrows()):
        if loop_index % (1000) == 0:
            logger.info(loop_index)
        try:
            sm = sa.compute_document_second_moment_with_tf([row['review']], corpus_mean, min_freq = 1)  
        except ValueError as ve:
#             logger.exception("no vocab")
            continue
        except:
            logger.exception("???")
            continue

        fo_k.write("{}\t{}\t{}\t{}\n".format(row['genre'], row['rating'], row['review'], "\t".join([str(v) for v in sm[0]])))
#         break
```
